```python
import pandas as pd
import json

df = pd.read_excel('Articles.xlsx', sheet_name='Articles')

df
```

```python
articles_to_pass = ["A Video Is Worth 4096 Tokens: Verbalize Videos To Understand Them In Zero Shot", "UPB at ACTI: Detecting Conspiracies using fine tuned Sentence Transformers"]
articles = {}
for _, row in df.iterrows():
    key = row[df.columns[0]]
    info = {}
    if key in articles_to_pass:
        continue
    for col in df.columns[1:]:
        if col == "PDF":
            break
        info[col] = row[col]
    articles[key] = info
    
json_file = json.dumps(articles, indent=4)
print(json_file)
open("articles.json", "w").write(json_file)
    
```
